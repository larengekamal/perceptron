# Perceptron - Training

### Importing Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating Dataset
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                          n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
y.shape
```

```python
# Scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

### Weigth Updation Formula

$\Large w_{new}=w_{old}+\eta*(y_i-\widehat{y_i})*X_i$

$w_{new}$ = New updated weight

$w_{old}$ = Old weight

$\eta$ = Learning Rate

$y_i$ = Output from Dataset

$\widehat{y_i}$ = Output from Model

$X_i$ = Input

```python
# Step Function
def step(z):
    return 1 if z>0 else 0
```

```python
# Perceptron Function
def Perceptron(X,y,learning_rate=0.01, epochs=1000):
    print(f"Shape of X={X.shape[0]},{X.shape[1]}")
    print(f"Shape of y={y.shape[0]}")
    X = np.insert(X,0,1, axis=1)
    print(f"Adding Column with Value 1 to X: {X}")
    weights = np.ones(X.shape[1])
    print(f"Creating weights with Value 1: {weights}")   
    
    for i in range(epochs):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + learning_rate*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]    
```

```python
intercept_,coef_ = Perceptron(X,y)
print(f"Coefficient={coef_}")
print(f"Interception={intercept_}")
```

```python
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

## Animation

```python
# Perceptron Function
def Perceptron_animation(X,y,learning_rate=0.01, epochs=1000):
    m=[]
    b=[]
    print(f"Shape of X={X.shape[0]},{X.shape[1]}")
    print(f"Shape of y={y.shape[0]}")
    X = np.insert(X,0,1, axis=1)
    print(f"Adding Column with Value 1 to X: {X}")
    weights = np.ones(X.shape[1])
    print(f"Creating weights with Value 1: {weights}")   
    
    for i in range(epochs):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + learning_rate*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
m,b = Perceptron_animation(X,y)
```

```python
%matplotlib notebook
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
fig, ax=plt.subplots(figsize=(10,5))

x_i=np.arange(-3,3,0.1)
y_i=m[0]*x_i+b[0]

ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
line,=ax.plot(x_i,y_i,'r-',linewidth=3)
plt.ylim(-3,3)

def update(i):
    label='epoch={0}'.format(i+1)
    line.set_ydata(m[i]*x_i+b[i])
    ax.set_xlabel(label)
    
anim=FuncAnimation(fig,update,repeat=False, frames=1000,interval=100)
```
